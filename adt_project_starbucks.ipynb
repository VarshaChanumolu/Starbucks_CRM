How many members belong to each gender category?

```python
c.execute('''select gender,count(*) as cnt from profile group by gender order by cnt desc''')
df = pd.DataFrame(c.fetchall(), columns=['Gender', 'Count'])
display(df)
```

What is the average income of members by gender?

```python
c.execute('''select gender,avg(income) from profile group by gender''')
df = pd.DataFrame(c.fetchall(), columns=['Gender', 'Average Income'])
display(df)
```

What is the age range of members in the dataset?

```python
c.execute('''select min(age) as min_age,max(age) as max_age from profile''')
df = pd.DataFrame(c.fetchall(), columns=['Min Age', 'Max Age'])
display(df)
```

How many members joined each year?


```python
c.execute('''SELECT SUBSTRING(CAST(became_member_on AS TEXT), 1, 4) AS join_year, COUNT(*) AS member_count
FROM profile
GROUP BY join_year
ORDER BY join_year;''')
df = pd.DataFrame(c.fetchall(), columns=['Join Year', 'Member Count'])
display(df)
```

Which id values are duplicated in the dataset?

```python
c.execute('''Select id,count(id) as cnt from profile group by id having cnt>1''')
df = pd.DataFrame(c.fetchall(), columns=['ID', 'Count'])
display(df)
```

 How many members are aged 50 and above?

```python
c.execute('''select count(*) as cnt from profile where age>=50''')
df = pd.DataFrame(c.fetchall(), columns=['Count'])
display(df)
```

What is the maximum income for each gender?

```python
c.execute('''select gender,max(income) from profile group by gender''')
df = pd.DataFrame(c.fetchall(), columns=['Gender', 'Max Income'])
display(df)
```

 What is the average reward for each offer_type?

```python
c.execute('''select offer_type,avg(reward) from portfolio
group by offer_type''')
df = pd.DataFrame(c.fetchall(), columns=['Offer_Type', 'Average Reward'])
display(df)
```

Which channels are most frequently used across all offers?

```python
c.execute('''Select channels ,count(*) as cnt from portfolio group by channels order by cnt desc''')
df = pd.DataFrame(c.fetchall(), columns=['Channels', 'Count'])
display(df)
```

Which offers include social as one of their channels?

```python
c.execute('''select offer_type,channels from portfolio where channels like '%social%' ''')
df = pd.DataFrame(c.fetchall(), columns=['Offer_Type', 'Channels'])
display(df)
```

How many events of each type are in the dataset?

```python
c.execute('''select event,count(*) as cnt from transcript group by event order by cnt desc''')
df = pd.DataFrame(c.fetchall(), columns=['Event', 'Count'])
display(df)
```

How many unique person IDs exist in the dataset?

```python
c.execute('''select count(distinct(person)) as cnt from transcript''')
df = pd.DataFrame(c.fetchall(), columns=['Count'])
display(df)
```

What is the distribution of events over time?

```python
c.execute('''Select event,count(*) as cnt from transcript group by time order by time desc''')
df = pd.DataFrame(c.fetchall(), columns=['Event', 'Count'])
display(df)
```

What percentage of events are offer viewed vs offer received?

```python
c.execute('''SELECT event, COUNT(*) * 100.0 / (SELECT COUNT(*) FROM transcript) AS percentage
FROM transcript
WHERE event IN ('offer viewed', 'offer received')
GROUP BY event;''')
df = pd.DataFrame(c.fetchall(), columns=['Event', 'Count'])
display(df)
```

What is the total number of offers received by each person?

```python
c.execute('''select person,count(*) as cnt from transcript where event='offer received' group by person''')
df = pd.DataFrame(c.fetchall(), columns=['Person', 'Count'])
display(df)
```

What is the average income of members who completed at least one offer?

```python
c.execute('''select avg(income) from profile where id in (select person from transcript where event='offer completed')''')
df = pd.DataFrame(c.fetchall(), columns=['Average Income'])
display(df)
```
